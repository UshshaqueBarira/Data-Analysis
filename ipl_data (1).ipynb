```python
ipl['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="b")
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
plt.figure(figsize=(8,5))
plt.hist(batting_first['win_by_runs'])
plt.title('Winning by Runs')
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
         
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=['yellow','blue','orange'])
plt.show()
```

```python
#numerical, categorical, condition of colors
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()[0:5]),labels=list(batting_first['winner'].value_counts()[0:5].keys()),autopct='%0.1f%%')
plt.show()
```

```python
batting_second=ipl[ipl['win_by_wickets']!=0]
batting_second.head()
```

```python
plt.figure(figsize=(6,6))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(list(batting_second['winner'].value_counts()[0:5]),labels=list(batting_second['winner'].value_counts()[0:5].keys()),autopct='%1.0f%%')
plt.show()
```
